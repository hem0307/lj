## HEATMAP

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.DataFrame({"maths":[80,70,60,40],"sci":[76,90,45,64],"eng":[87,81,73,61],"hindi":[74,65,45,82]})
sns.heatmap(df)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.DataFrame({"maths":[80,70,60,40],"sci":[76,90,45,64],"eng":[87,81,73,61],"hindi":[74,65,45,82]},index=["A","B","C","D"])
sns.heatmap(df,vmin=40,vmax=100,cmap="YlGnBu",linewidths=1,linecolor="white",annot=True,cbar=True)
plt.show()
```

## Boxplot

```python
df=pd.DataFrame({"dept":["IT","IT","HR","HR","Finance","IT","HR"],
                 "gender":["M","M","F","M","F","F","M"],
                 "Salary":[50000,60000,45000,48000,70000,55000,45000],})
sns.boxplot(x="dept",y="gender",data=df,color="red",hue="gender")
plt.show()
```

```python
df=pd.DataFrame({"dept":["IT","IT","HR","HR","Finance","IT","HR"],
                 "gender":["M","M","F","M","F","F","M"],
                 "Salary":[50000,60000,45000,48000,70000,55000,45000],})
sns.scatterplot(x="dept",y="Salary",data=df,color="red",hue="gender")
plt.show()
```
