```python
import pandas as pd
em=pd.DataFrame({"emp_id":[1,2,3,4],"name":["A","B","C","D"],"dept_id":[10,20,30,40]})
print(em)
print("------------------------------>")

dept=pd.DataFrame({"dept_id":[10,20,30],"dept_name":["HR","IT","Finance"]})
print(dept)
print("------------------------------>")

pd.merge(em,dept)
```

```python
pd.merge(em,dept,on="dept_id",how="inner")
```

```python
pd.merge(em,dept,on="dept_id",how="outer")
```

```python
pd.merge(em,dept,on="dept_id",how="left")
```

```python
pd.merge(em,dept,on="dept_id",how="right")
```

```python
import pandas as pd
em=pd.DataFrame({"emp_id":[1,2,3,4],"name":["A","B","C","D"],"d_id":[10,20,30,40]})
print(em)
print("------------------------------>")

dept=pd.DataFrame({"dept_id":[10,20,30],"dept_name":["HR","IT","Finance"]})
print(dept)
print("------------------------------>")

pd.merge(em,dept,left_on="d_id",right_on="dept_id")
```

```python
import pandas as pd
em=pd.DataFrame({"emp_id":[1,2,3,4],"name":["A","B","C","D"],"d_id":[10,20,30,40]})
print(em)
print("------------------------------>")

dept=pd.DataFrame({"dept_id":[10,40,20],"dept_name":["HR","IT","Finance"]})
print(dept)
print("------------------------------>")

pd.merge(em,dept,left_index=True,right_index=True)
```

```python
import pandas as pd
em=pd.DataFrame({"emp_id":[1,2,3,4],"name":["A","B","C","D"],"dept_id":[10,20,30,40],"updated":[1,2,3,4]})
print(em)
print("------------------------------>")

dept=pd.DataFrame({"dept_id":[10,20,30],"dept_name":["HR","IT","Finance"],"updated":[1,2,3]})
print(dept)
print("------------------------------>")

pd.merge(em,dept,on="dept_id")
pd.merge(em,dept,on="dept_id",suffixes=("_emp","_dept"))
```

groupby()

```python
import pandas as pd
df=pd.DataFrame({"dept":["IT","IT","HR","HR","FINANCE","IT","HR"],
                "emp":["A","B","C","D","E","F","C"],
                "salary":[50000,40000,60000,75000,55000,30000,48000],
                "exp":[2,3,1,2,5,4,1]})

print(df)
print("------------------------------>")

df.groupby("dept").sum(numeric_only=True)
```

```python
df.groupby("dept")["salary"].sum(numeric_only=True)
```

```python
df.groupby("dept").mean(numeric_only=True)
```

```python
print(df.groupby(["dept","exp"]).sum(numeric_only=True))
```

```python
print(df.groupby("dept",sort=False).sum(numeric_only=True))
```

```python
df.groupby("dept").agg({"salary":["min","max","mean"],"exp":["mean"]})
```

```python
df.groupby("dept").groups
```

```python
df.groupby("dept").describe()
```

```python
df.groupby("dept").nunique()
```

```python
df.groupby("dept")["salary"].apply(lambda x:x.max()-x.min())
```

```python
df.groupby("dept").nth(0)
```

```python
df.groupby("dept").nth(-1)
```

```python
df.groupby("dept").nth([0,-1])
```

```python
def label_high(g):
    g["high_earner"]=g["salary"]>=g["salary"].mean()
    return g

res=df.groupby("dept").apply(label_high)
res
```
