## Networkx

```python
import networkx as nx

g=nx.Graph()
g.add_node(1)
g.add_nodes_from([2,3,4])
g.add_nodes_from(range(5,7))
g.add_edge(1,2)
g.add_edges_from([(1,3),(3,4),(3,1),(4,6),(5,3),(5,6)])
nx.draw_networkx(g,node_size=850,node_color="yellow",width=4,edge_color="red")
```
