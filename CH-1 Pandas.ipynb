# Series()

```python
import pandas as pd
data=['ABC','PQR','XYZ']
ds=pd.Series(data)
print(ds)
print(type(ds))

ds1=pd.Series(data,index=['A','B','C'])
print(ds1)

d={'A':10,'B':20,'C':30}
ds2=pd.Series(d)
print(ds2)

ds3=pd.Series(range(51,61))
print(ds3)
print(type(ds3))
```

# DataFrame()

```python
import pandas as pd

data={'name':['Roy','Harsh','Om','Veer'],
      'div':['D1','D2','D3','D4'],
      'marks':[30,40,28,32]}
df=pd.DataFrame(data)
print(df)
print(type(df))

d=[['A',20,'D1'],
   ['B',30,'D2'],
   ['C',40,'D3']]
df1=pd.DataFrame(d,columns=['name','marks','div'],index=['AA','BB','CC'])
print(df1)
```

```python
import pandas as pd

s1=pd.Series([20,21,22])
s2=pd.Series([10,12,12])
s3=pd.Series([15,25,22])

df=pd.DataFrame({'col1':s1,'col2':s2})
print(df)

d=[{'A':20,'B':30},{'A':21,'B':35},{'A':25,'B':35}]
df2=pd.DataFrame(d)
print(df2)
```

# Sorting

```python
import pandas as pd

data={'name':['Rohit','Hahul','Gill','Kohli','Rohit','Kohli'],
      'runs':[12000,1000,6000,4000,12000,4000],
      'avg':[59.5,48.7,51.2,49.3,59.5,49.3]}
df=pd.DataFrame(data)
print(df)

df.sort_values(by='runs')
```

```python
df.sort_values(by='runs',ascending=False)
```

```python
df.sort_index(ascending=False)
```

```python
df.sort_index(axis=1)
```

```python
df[df['runs']>5000]
```

```python
df[(df['runs']>5000) & (df['avg']<30.0)]
```
