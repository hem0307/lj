## 12

```python
#12
age_group=[]
for i in df["Age"]:
    if i<=18:
        age_group.append("Child :")
    elif i<=30:
        age_group.append("Young :")
    elif i<=50:
        age_group.append("Adult :")
    else:
        age_group.append("Senior :")


df["Age_group"]=age_group
pd.crosstab(df["Age_group"],df["Survived"])
```
